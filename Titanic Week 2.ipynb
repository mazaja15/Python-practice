```python
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
```

```python
binlabels = ['Child', 'Teen', 'Twenties', 'Thirties', 'Forties', 'Fifties', 'Sixties', 'Seventies', 'Eighties', 'Nighties']
```
