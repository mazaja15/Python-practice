```python
wordcount = dict()
```

```python
words = []
with open(r'c:\Data\car.txt') as c:
    for line in c:
        for i in line.split():
            words.append(i)
            wordcount[i] = wordcount.get(i,0) + 1
```

```python
wordcount
```

```python
#  since we cannot sort dictionary, this is the way to do it. The x[1] is telling us by what index position we want to sort it by. If we did x[0] it would sort by the word not number.
sorted(wordcount.items(), key=lambda x: x[1], reverse=True)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop = set(stopwords.words('english')) 
```

```python
stop
```

```python
wordcount = {}
with open(r'c:\Data\car.txt') as c:
    for line in c:
        for i in line.split():
            if i.replace('\'','').lower() not in stop:
                wordcount[i] = wordcount.get(i,0) + 1
```

```python
sorted(wordcount.items(), key=lambda x: x[1], reverse=True)
```
